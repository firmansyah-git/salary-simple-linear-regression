## 1. Import Libraries  
Import pustaka yang digunakan dalam analisis ini. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
```

## 2. Load Dataset
Dataset dimuat menggunakan `pandas` dan disimpan dalam DataFrame `df`.

```python
df = pd.read_csv("Salary_dataset.csv", index_col=0)
df
```

## 3. Exploratory Data Analysis
### Years of Experience and Salary Distribution
Visualisasi distribusi data dari `YearsExperience` dan `Salary`


```python
for column in df.columns:
    plt.hist(df[column], color='skyblue', edgecolor='black')
    plt.title(f'{column} Distributions')
    plt.ylabel(f'Frequency of column')
    plt.show()
```

```python
sns.scatterplot(data=df, x="YearsExperience", y="Salary")
plt.title("Years of experience causes salary")
plt.show()
```

## 4. Data Modeling

### Memisahkan variabel untuk fitur(`X`) and target(`y`)

```python
X = df[["YearsExperience"]]
y = df["Salary"]
```

### Initialitation and Model Training

```python
model = LinearRegression()
model.fit(X, y)
```

### Mode Interpretation

```python
print("Intercept (bias):", model.intercept_)
print("Slope (koefisien):", model.coef_[0])
```

Berdasarkan hasil regresi linier, setiap tambahan 1 tahun pengalaman kerja diperkirakan akan meningkatkan gaji sebesar **9,449.96**, dimulai dari nilai dasar sebesar **24,848.20** ketika pengalaman kerja adalah nol.

**Persamaan model:**  
Salary = 9,449.96 Ã— YearsExperience + 24,848.20

### Prediction

```python
prediction = model.predict(pd.DataFrame([[8.5]], columns=["YearsExperience"]))
print(prediction[0])
```

ðŸ” Prediksi Gaji untuk 8.5 tahun pengalaman kerja:  
ðŸ’° Diperkirakan: $ 105,172.88 

### Visualitation of Linear Regression

```python
plt.scatter(X, y, color="blue", label="Data Aktual")
plt.plot(X, model.predict(X), color="red", label="Regresi Linear")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.legend()
plt.title("Linear Regression: Years Experience vs Salary")
plt.grid(True)
plt.show()
```

### Model Evaluation

```python
y_pred = model.predict(X)

print("MSE:", mean_squared_error(y, y_pred))
print("RÂ² Score:", r2_score(y, y_pred))
```

Setelah membangun model regresi linear untuk memprediksi gaji berdasarkan jumlah tahun pengalaman kerja, berikut adalah hasil evaluasi performa model:

- **MSE (Mean Squared Error)**: 31,270,951.72
MSE mengukur rata-rata kuadrat dari selisih antara nilai aktual dan prediksi. Nilai ini menunjukkan bahwa secara rata-rata, kuadrat error per prediksi cukup besar, namun sulit diinterpretasikan secara langsung karena bukan dalam satuan gaji.

- **RMSE (Root Mean Squared Error)**: Â±5,591,147
RMSE lebih intuitif karena memiliki satuan yang sama dengan target (gaji). Artinya, model memprediksi gaji dengan rata-rata kesalahan sekitar Â±5.59 juta rupiah dibanding nilai aktual.

- **Relative Error**: ~8%
Relative error dihitung dengan membagi RMSE dengan rata-rata gaji (70 juta). Ini menunjukkan bahwa model melakukan prediksi dengan rata-rata kesalahan hanya sekitar 8% dari gaji rata-rata, yang tergolong sangat baik untuk skenario regresi di dunia nyata.

ðŸ“ˆ RÂ² Score: 0.957
Model mampu menjelaskan sekitar 95.7% variabilitas data gaji berdasarkan lama pengalaman kerja, yang merupakan indikasi bahwa model sangat fit terhadap data.
